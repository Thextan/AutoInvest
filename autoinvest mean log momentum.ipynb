```python
import pandas as pd
import numpy as np
df = pd.read_json(r"/tmp/EUR_USD.M15.json")
df = df.set_index('time')
```

```python
df
```

```python
results = []
for i in df.iterrows():
#     time, mid = i[0], i[1]['mid']
    mid = i[1]['mid']
    close, open = mid['c'], mid['o']
    results.append((i[0],close,open))
for j in results:
    print(j)
```

```python
df['close'] = [float(result[1]) for result in results]
df['open'] = [float(result[2]) for result in results]
df['c-returns'] = np.log(df['close'] / df['close'].shift(1))
df['o-returns'] = np.log(df['open'] / df['open'].shift(1))
df['c-returns']
```

```python
cols = []

for momentum in [15, 30, 120]:
    col = 'position_%s' % momentum
    df[col] = np.sign(df['c-returns'].rolling(momentum).mean())
    cols.append(col)
```

```python
%matplotlib inline
import seaborn as sns; sns.set()

stratsc = ['c-returns']

for col in cols:
    strat = 'strategy_%s' % col.split('_')[1]
    df[strat] = df[col].shift(1) * df['c-returns']
    stratsc.append(strat)

df[stratsc].dropna().cumsum().apply(np.exp).plot()
```

```python
for momentum in [15, 30, 120]:
    col = 'position_%s' % momentum
    df[col] = np.sign(df['o-returns'].rolling(momentum).mean())
    cols.append(col)
stratso = ['o-returns']
for col in cols:
    strat = 'strategy_%s' % col.split('_')[1]
    df[strat] = df[col].shift(1) * df['o-returns']
    stratso.append(strat)
df[stratso].dropna().cumsum().apply(np.exp).plot()
```
